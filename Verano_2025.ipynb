```python
def abrir_info(path):
  try:
    with open(path, 'r', encoding='latin-1') as archivo:
        print("\nContenido del archivo leido correctamente")
        contenido = archivo.readlines()
        contenido=contenido[17:]
        #content_after_skip = "".join(contenido[17:])
        #formatted_content = "Spectrometer Data\n" + content_after_skip.strip()
        return contenido

  except FileNotFoundError:
    print("Error: El archivo"+ path+"no se encontró.")
  except Exception as e:
    print(f"Ocurrió un error inesperado: {e}")
```

```python
import matplotlib.pyplot as plt

def parse_data_for_plotting(data_list_of_strings):
    parsed_data = []
    for item_str in data_list_of_strings:
        parts = item_str.strip().split('\t')
        if len(parts) == 2:
            try:
                wavelength = float(parts[0])
                reflection = float(parts[1])
                parsed_data.append({'wavelength': wavelength, 'reflection': reflection})
            except ValueError:
                print(f" No se pudo convertir '{item_str.strip()}' a numeros. saltado.")
        else:
            print(f"formato inesperado '{item_str.strip()}'. saltado.")
    return parsed_data
```

```python
from google.colab import drive
import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from scipy.signal import savgol_filter
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler


drive.mount('/content/drive')
base_path = "/content/drive/MyDrive/Verano UG 2025/Mediciones 2/"
subfolders = [
    "06-02-2025", "061-02-2025", "062-02-2025", "063-02-2025",
    "07-02-2025", "071-02-2025", "072-02-2025", "073-02-2025",
    "074-02-2025", "075-02-2025", "08-02-2025", "081-02-2025",
    "082-02-2025", "083-02-2025", "09-02-2025", "091-02-2025",
    "092-02-2025", "093-02-2025"
]
prefixes = ['hojaenferma', 'hojasana', 'hojaseca']
suffixes = ['a', 'b', 'c', 'd', 'a 2', 'b 2', 'c 2', 'd 2']

# ======================
# FUNCIONES
#
def load_data(file_path):
    #Carga datos omitiendo las primeras 17 líneas
    try:
        with open(file_path, 'r', encoding='latin-1') as f:
            return f.readlines()[17:]
    except Exception as e:
        print(f"Error leyendo {file_path}: {str(e)}")
        return None

def parse_data(content):
    #Convierte contenido en datos estructurados
    if not content:
        return []

    parsed = []
    for line in content:
        parts = line.strip().split()
        if len(parts) >= 2:
            try:
                parsed.append({
                    'wavelength': float(parts[0]),
                    'reflection': float(parts[1])
                })
            except ValueError:
                continue
    return parsed

def plot_spectrum(data, title, ax=None, color='b'):
    #Grafica espectros de reflexión
    if not data:
        return

    if ax is None:
        fig, ax = plt.subplots(figsize=(10, 6))

    wavelengths = [d['wavelength'] for d in data]
    reflections = [d['reflection'] for d in data]

    ax.plot(wavelengths, reflections, color=color)
    ax.set_title(title)
    ax.set_xlabel('Longitud de Onda (nm)')
    ax.set_ylabel('Reflexión (%)')
    ax.grid(True)
    return ax

def smooth_data(data, method='savgol', window=21, order=2):
    #Aplica suavizado a los datos
    if not data:
        return []

    reflections = np.array([d['reflection'] for d in data])

    if method == 'savgol':
        smoothed = savgol_filter(reflections, window, order)
    else:  # media móvil
        smoothed = np.convolve(reflections, np.ones(window)/window, mode='same')

    return [{
        'wavelength': data[i]['wavelength'],
        'reflection': smoothed[i]
    } for i in range(len(data))]

def apply_pca(data_list, n_components=2):
    #Aplica Análisis de Componentes Principales
    # Encontrar la longitud mínima para truncar
    min_len = min(len(d) for d in data_list if d)

    # Crear matriz de características
    X = np.zeros((len(data_list), min_len))
    for i, data in enumerate(data_list):
        if data:
            X[i] = [d['reflection'] for d in data[:min_len]]

    # Estandarizar y aplicar PCA
    X_std = StandardScaler().fit_transform(X)
    pca = PCA(n_components=n_components)
    return pca.fit_transform(X_std), pca

# ======================
# PROCESAMIENTO PRINCIPAL
# ======================
# Recopilar y procesar todos los archivos
all_data = []
all_smoothed = []
file_info = []

for folder in subfolders:
    folder_path = os.path.join(base_path, folder)
    if not os.path.exists(folder_path):
        continue

    for file in os.listdir(folder_path):
        if not file.endswith('.txt'):
            continue

        # Determinar tipo de hoja
        file_type = None
        for prefix in prefixes:
            if file.startswith(prefix):
                file_type = prefix.replace('hoja', '')
                break

        if not file_type:
            continue

        file_path = os.path.join(folder_path, file)
        content = load_data(file_path)
        if not content:
            continue

        # Parsear datos
        parsed = parse_data(content)
        if not parsed:
            continue

        # Almacenar información
        file_info.append({
            'file': file,
            'type': file_type,
            'folder': folder,
            'path': file_path
        })
        all_data.append(parsed)

        # Suavizar datos
        smoothed = smooth_data(parsed)
        all_smoothed.append(smoothed)

# ======================
# VISUALIZACIÓN
# ======================
# Graficar espectros originales vs suavizados
fig, axs = plt.subplots(3, 1, figsize=(12, 15), sharex=True)
colors = {'enferma': 'red', 'sana': 'green', 'seca': 'blue'}

for i, file_type in enumerate(['enferma', 'sana', 'seca']):
    # Seleccionar datos para este tipo
    indices = [idx for idx, info in enumerate(file_info) if info['type'] == file_type]

    # Graficar primeros 3 ejemplos
    for j, idx in enumerate(indices[:3]):
        color = colors[file_type]
        label_orig = f"{file_type.capitalize()} Original" if j == 0 else None
        label_smooth = f"{file_type.capitalize()} Suavizado" if j == 0 else None

        # Datos originales
        plot_spectrum(
            all_data[idx],
            f'Espectros {file_type.capitalize()}',
            ax=axs[i],
            color=color if j==0 else None
        )

        # Datos suavizados
        plot_spectrum(
            all_smoothed[idx],
            f'Espectros {file_type.capitalize()}',
            ax=axs[i],
            color='purple' if j==0 else None
        )

    axs[i].legend(['Original', 'Suavizado'])

plt.tight_layout()
plt.savefig('comparacion_espectros.png')
plt.show()

# ======================
# ANÁLISIS PCA
# ======================
# Aplicar PCA
pca_result, pca_model = apply_pca(all_smoothed)

# Crear DataFrame con resultados PCA
pca_df = pd.DataFrame({
    'PC1': pca_result[:, 0],
    'PC2': pca_result[:, 1],
    'Tipo': [info['type'] for info in file_info],
    'Archivo': [info['file'] for info in file_info],
    'Carpeta': [info['folder'] for info in file_info]
})

# Guardar resultados PCA
pca_df.to_csv('resultados_pca.csv', index=False)

# Graficar resultados PCA
plt.figure(figsize=(10, 8))
for file_type, color in colors.items():
    subset = pca_df[pca_df['Tipo'] == file_type]
    plt.scatter(subset['PC1'], subset['PC2'],
                c=color, label=file_type.capitalize(),
                alpha=0.6, edgecolors='w', s=100)

plt.title('Análisis PCA de Espectros de Hojas')
plt.xlabel('Componente Principal 1')
plt.ylabel('Componente Principal 2')
plt.legend()
plt.grid(True)
plt.savefig('resultado_pca.png')
plt.show()

print(f"Archivos procesados: {len(file_info)}")
print("Nombre archivos:")
print("comparacion_espectros.png")
print("resultado_pca.png")
print("resultados_pca.csv")
```
